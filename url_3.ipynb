```python
import pandas as pd
```

```python
# import datasets
#internal_all = pd.read_csv("internal_all.csv", encoding='unicode_escape')
urls = pd.read_csv("urls.csv")
valid_urls = pd.read_csv("valid_urls.csv")

valid_urls.head(5)
```

```python
urls.head(5)
```

```python
# filter for urls 
urls['QUERY_COUNT'] = urls['QUERY'].str.count(',') + 1
filtered_urls = urls[urls['QUERY_COUNT'] > 0]
filtered_urls = filtered_urls.dropna()
final_urls = pd.merge(filtered_urls, valid_urls, left_on = "URL", right_on = "loc", how = "inner")
final_urls = final_urls.drop(columns=['loc', 'changefreq', 'priority', 'lastmod'])
final_urls.head()
```

```python
final_urls.to_csv("final_urls.csv")
```
