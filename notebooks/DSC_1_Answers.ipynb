__Question 3:__ What is the average salary for someone who holds a Bacherlor's Degree and is employed? 

__Bonus Question 1:__ Find the average for 0-2 years, 3-5 years, and 5-10 years of experience.

```python
# Syntax to know: filtering the df: (df[df['column'] > conditional]),
#                 grouping: df.groupby(),
#                 average: df.mean()

# Write your answer here:
bachelors = salaries[salaries['EdLevel'] == 'Bachelor’s degree (B.A., B.S., B.Eng., etc.)']
avg_comp = bachelors[['ConvertedCompYearly']].mean()
avg_comp

bachelors = salaries[salaries['EdLevel'] == 'Bachelor’s degree (B.A., B.S., B.Eng., etc.)']
employed = bachelors[bachelors['Employment'] == 'Employed']
yoe = employed[['WorkExp', 'ConvertedCompYearly']].groupby('WorkExp').mean().head(10)
yoe
```
